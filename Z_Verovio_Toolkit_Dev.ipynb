```python
import intervals
from intervals import * 
from intervals import main_objs
import pandas as pd
import re
from ipywidgets import interact
from IPython.display import display
import requests
import os
import verovio
from IPython.display import SVG
from IPython.display import display, HTML
from warnings import *
```

### Import Piece and Set Up Verovio 

```python
# Select a prefix:
# prefix = 'Music_Files/'
prefix = 'https://crimproject.org/mei/'
# Add your filename here
mei_file = 'CRIM_Model_0008.mei'
url = prefix + mei_file
piece = importScore(url)
print(piece.metadata)
response = requests.get(url)
fetched_mei_string = response.text
tk = verovio.toolkit()
tk.loadData(fetched_mei_string)
tk.setScale(30)

tk.setOption( "pageHeight", "1000" )
tk.setOption( "pageWidth", "3000" )
```

### Show Cadences as Verovio Clips

- These are three-voice segments
- The end-point of the cadence is the middle measure

```python
# find the cadences
cads = piece.cadences()
cad_list = cads["Measure"].tolist()
cads.fillna("Unknown", inplace=True)

# get the cadence metadata for caption and make dictionary
for cad in cad_list:
    cad_index = cad_list.index(cad)
    c_tone = cads.iloc[cad_index]["Tone"]
    c_type = cads.iloc[cad_index]["CadType"]
    c_beat = cads.iloc[cad_index]["Beat"]
    evaded_c = cads.loc[cads['Measure'] == cad, 'Evaded'].item()
    low = cad-1
    high = cad+1
    mr = str(low) + "-" + str(high)
    mdict = {'measureRange': mr}
    
    # select verovio measures and redo layout
    tk.select(str(mdict))
    tk.redoLayout()
    
    # get the number of pages and display the music
    count = tk.getPageCount()
    for c in range(1, count + 1):
        music = tk.renderToSVG(c)
        print("MEI File: ", mei_file)
        print(piece.metadata['composer'])
        print(piece.metadata['title'])
        print("Cadence End Measure:", cad)
        print("Beat: ", c_beat)
        print("Cadence Tone: ", c_tone)
        print("Cadence Type: ", c_type)
        print("Evaded Cadence: ", evaded_c)
        display(SVG(music))
    
```

### Show Vervio Clips for Presentation Types

- The measures shown start with the first voice of the first entry, and extend to four bars after the last entry

```python
p_types = piece.presentationTypes(limit_to_entries = True,
                        edit_distance_threshold = 0,
                        include_hidden_types = False,
                        combine_unisons = False,
                       melodic_ngram_length = 4)
```

```python
print("Results:")
# print("Limit to Entries: ", limit_to_entries)
# print("Edit Distance Threshold: ", edit_distance_threshold)
# print("Include Hidden Types: ", include_hidden_types)
# print("Combine Unisons: ", combine_unisons)
# print("Melodic Ngram Length: ", melodic_ngram_length)

# collect the metadata
for p_type in p_types.index:
    this_p_type = p_types.iloc[p_type]["Presentation_Type"]
    p_voices = p_types.iloc[p_type]["Voices"]
    n_voices = p_types.iloc[p_type]["Number_Entries"]
    soggetto = p_types.iloc[p_type]["Soggetti"][0]
    mint = p_types.iloc[p_type]["Melodic_Entry_Intervals"]
    tint = p_types.iloc[p_type]["Time_Entry_Intervals"]
    flexed = p_types.iloc[p_type]["Flexed_Entries"]
    ml = p_types.iloc[p_type]["Measures_Beats"]
    
    # build the measure range dictionary
    first = ml[0].split('/')[0]
    last = str(int(ml[-1].split('/')[0]) + 4)
    mr = str(first) + "-" + str(last)
    mdict = {'measureRange': mr}
    
    # select measures in verovio and redo the layout
    tk.select(str(mdict))
    tk.redoLayout()
    # get the number of pages
    count = tk.getPageCount()
    
    # print caption
    print("MEI File: ", mei_file)
    print(piece.metadata['composer'])
    print(piece.metadata['title'])
    print("Measures:", mr)
    print("Presentation Type: ", this_p_type)
    print("Voices: ", p_voices)
    print("Number of Entries: ", n_voices)
    print("Main Soggetto: ", soggetto)
    print("Melodic Entry Intervals: ", mint)
    print("Time Entry Intervals: ", tint )
    print("Flexed: ", flexed)
    # print the music
    for c in range(1, count + 1):
        music = tk.renderToSVG(c) 
        display(SVG(music))
```
