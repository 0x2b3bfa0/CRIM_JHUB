```python
def prRed(skk): print("\033[91m {}\033[00m" .format(skk))
def prGreen(skk): print("\033[92m {}\033[00m" .format(skk))
def prYellow(skk): print("\033[93m {}\033[00m" .format(skk))
def prLightPurple(skk): print("\033[94m {}\033[00m" .format(skk))
def prPurple(skk): print("\033[95m {}\033[00m" .format(skk))
def prCyan(skk): print("\033[96m {}\033[00m" .format(skk))
def prLightGray(skk): print("\033[97m {}\033[00m" .format(skk))
def prBlack(skk): print("\033[98m {}\033[00m" .format(skk))
from IPython import display
from IPython.display import display, Image
prefix = 'Exercises/Ex01_1_'
extension = '.png'
while True:
    interval = int(input('Enter any diatonic interval between 12 and -12: '))
    if interval == int('1'):
        prGreen('Unison:')
        name = str(interval)
        display(Image(filename=(prefix + name + extension), width=150))
        continue
    elif interval == int('2'):
        prGreen('Ascending ' + str(interval) + 'nd:')
        name = str(interval)
        display(Image(filename=(prefix + name + extension), width=150))
        continue
    elif interval == int('3'):
        prGreen('Ascending ' + str(interval) + 'rd:')
        name = str(interval)
        display(Image(filename=(prefix + name + extension), width=150))
        continue
    elif int('3') < interval < int('13'):
        prGreen('Ascending ' + str(interval) + 'th:')
        name = str(interval)
        display(Image(filename=(prefix + name + extension), width=150))
        continue
    elif int('12') < interval < int('999'):
        prRed('We are only considering intervals between -12 and 12')
        continue
    elif interval == int('0'):
        prRed('0 does not define an interval')
        continue
    elif interval == int('-1'):
        prRed('... almost any interval between 12 and -12')
        prRed('Unison is conventionally considered ascending, so -1 is not recognized.')
        continue
    elif interval == int('-2'):
        prGreen('Descending ' + str(abs(interval)) + 'nd:')
        name = str(interval)
        display(Image(filename=(prefix + name + extension), width=150))
        continue
    elif interval == int('-3'):
        prGreen('Descending ' + str(abs(interval)) + 'rd:')
        name = str(interval)
        display(Image(filename=(prefix + name + extension), width=150))
        continue
    elif int('-13') < interval < int('-3'):
        prGreen('Descending ' + str(abs(interval)) + 'th:')
        name = str(interval)
        display(Image(filename=(prefix + name + extension), width=150))
        continue
    elif interval < int('-12'):
        prRed('We are only considering intervals between -12 and 12')
        continue
    elif interval == int('999'):
        prCyan('***BRAVO! GO TO THE NEXT MODULE***')
        break
    else:
        prRed('Something went wrong')
        prRed('Make sure you are following the rules for entering diatonic intervals')
        continue
```
